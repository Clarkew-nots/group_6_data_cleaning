```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio


pio.renderers.default = "notebook"
```

```python
dataset = pd.read_csv("./data/Students_admission_records.csv")
dataset
```

```python
dataset.isna().sum()
```

```python
dataset["Age"].mask(dataset["Age"] < 1, np.nan)
```

```python
dataset.info()
```

```python
dataset["Admission Status"].count()
```

```python
accepted = dataset[dataset["Admission Status"] == "Accepted"]
rejected = dataset[dataset["Admission Status"] == "Rejected"]

accepted["High School Percentage"].min()
rejected["High School Percentage"].max()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(
    x=dataset["Admission Status"],
    y=dataset["Admission Test Score"]
)

)
fig.show()
```

```python
boxes = go.Figure()

boxes.add_trace(go.Box(x=dataset["Admission Test Score"], name="Admission Test Score"))
boxes.add_trace(go.Box(x=dataset["High School Percentage"], name="High School Percentage"))

boxes.show()
```

```python
percentage_median = dataset["High School Percentage"].median()
admission_median = dataset["Admission Test Score"].median()
```
